```python
from ggmap.readwrite import *
from ggmap.tree import *
from ggmap.utils import *
from ggmap.tree import _get_otus_from_clade
from ggmap.utils import _convert_metaphlan_profile_to_greengenes
```

```python
simlevel = '99'
```

```python
# read NCBI taxonomy
nodes = read_ncbi_nodes('NCBItaxonomy/nodes.dmp')
merged = read_ncbi_merged('NCBItaxonomy/merged.dmp')
tax = build_ncbi_tree(nodes, True)
```

```python
# read MetaPhlAn clade information
taxids_metaphlan = read_taxid_list('Cache/taxids_metaphlan.txt')
taxids_metaphlan = update_taxids(taxids_metaphlan, merged)
clades_metaphlan = read_metaphlan_markers_info('Metaphlan/markers_info.txt')
```

```python
# read GreenGenes information
gg_ids_accessions = read_gg_accessions('gg_13_5_accessions.txt')
gg_taxids = read_taxid_list('Cache/taxids_greengenes13_5.txt')
gg_taxids = update_taxids(gg_taxids, merged)
gg_otumap_orig = read_gg_otu_map('gg_13_5_otus/otus/%s_otu_map.txt' % simlevel,
                                 gg_ids_accessions)
```

```python
# tag NCBI taxonomy tree with GreenGenes OTUs and MetaPhlAn clades
tree_gg = map_onto_ncbi(tax, gg_otumap_orig, gg_taxids, 'otus', verbose=True)
```

```python
tree_mp = map_onto_ncbi(tax, clades_metaphlan, taxids_metaphlan, 'mp_clades',
                        verbose=True)
```

```python
# build map from MetaPhlAn clades to GreenGenes OTUs
resmap = match_metaphlan_greengenes(clades_metaphlan.keys(), tree_mp, 'mp_clades', tree_gg, 'otus')
```

```python
# write resulting map to a file
write_clade2otus_map('Cache/metaphlan_to_gg13_5-%s.txt' % simlevel, resmap)
```
